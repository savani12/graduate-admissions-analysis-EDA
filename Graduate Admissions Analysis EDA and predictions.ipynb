## Data Modelling

### 1. Simple Linear Regression

```python
X = df["GRE Score"].values
X=X/340
y= df["Chance of Admit "].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
print('Train size of X and Y: ',len(X_train),len(y_train))
```

```python
regressor.fit(X_train.reshape(-1,1), y_train)
```

```python
y_pred = regressor.predict(X_test.reshape(-1,1))
```

```python
regressor.score(X_test.reshape(-1,1),y_test.reshape(-1,1))
```

```python
plt.subplots(figsize=(12,8))
plt.scatter(X_train, y_train, color = "red")
plt.plot(X_train, regressor.predict(X_train.reshape(-1,1)), color = "green")
plt.title("GRE Scores vs Chance of Admit")
plt.xlabel("GRE Scores")
plt.ylabel("Chance Of Admit")
plt.show()
```

```python
plt.subplots(figsize=(12,8))
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train.reshape(-1,1)), color = 'blue')
plt.title('GRE Scores vs Chance of Admit (Test set)')
plt.xlabel('GRE Scores')
plt.ylabel('Chance of Admit')
plt.show()
```

### 3. Logistic Regression

```python
# If the "chance of admit" is greater than "80" then the admission is confirmed means "1"
y_train = [1 if each > 0.8 else 0 for each in y_train]
# If the "chance of admit" is lower than "80" then the admission is not confirmed means "0"
y_test = [1 if each > 0.8 else 0 for each in y_test]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
print(classifier.predict(sc.transform([[325, 100, 3, 4.1, 3.7, 7.67]])))
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred)*100))
```

### 4. Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 100, random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred)*100))
```
